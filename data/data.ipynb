The SLOG dataset is built on top of the COGS training set, which contains $24,155$ unique sentences. In theory, this should mean that all sentences from the original COGS should be present within SLOG (?); however, this does not seem to be the case when we take a look at the set difference:

```python
len(set(train_cogs["sent"]) - set(train_slog["sent"]))
```

We can take a look at ten of the sentences present in COGS that seem to be "missing" from SLOG:

```python
list(set(train_cogs["sent"]) - set(train_slog["sent"]))[:10]
```

Furthermore, the "missing" entries seem to affect only `in_distribution` examples:

```python
train_slog[train_slog["type"] == "in_distribution"]
```

```python
train_cogs[train_cogs["type"] == "in_distribution"]
```

```python
set1 = set(train_cogs[train_cogs["type"] == "in_distribution"]["sent"])
set2 = set(train_slog[train_slog["type"] == "in_distribution"]["sent"])

only_in_df1 = set1 - set2
only_in_df2 = set2 - set1

print("in_dist sents only in COGS:", len(only_in_df1))
print("in_dist sents only in SLOG:", len(only_in_df2))
```

So it seems that $1,155$ in-distribution sentences have been modified/removed between COGS and SLOG?
